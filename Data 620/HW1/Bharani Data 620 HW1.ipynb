In this assignment, you’re asked to create the nodes and edges for a basic graph, such as the Krackhardt kite shown below. (You’re welcome to substitute data of your own choosing).

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
%matplotlib inline

G=nx.Graph()

# create edges
G.add_edge('Tyler','John')
G.add_edge('Tyler','Euclid')
G.add_edge('Tyler','Stefano')
G.add_edge('Tyler','Mustafa')
G.add_edge('John','Tyler')
G.add_edge('John','Stefano')
G.add_edge('John','Bharani')
G.add_edge('John','Karim')
G.add_edge('Euclid','Tyler')
G.add_edge('Euclid','Stefano')
G.add_edge('Euclid','Mustafa')
G.add_edge('Mustafa','Karim')
G.add_edge('Mustafa','Stefano')
G.add_edge('Mustafa','Euclid')
G.add_edge('Karim','Bharani')
G.add_edge('Karim','Stefano')
G.add_edge('Karim','Mustafa')
G.add_edge('Karim','Tage')
G.add_edge('Bharani','John')
G.add_edge('Bharani','Stefano')
G.add_edge('Bharani','Karim')
G.add_edge('Tage','Mustafa')
G.add_edge('Tage','Karim')
G.add_edge('Tage','Ian')
G.add_edge('Ian','Tage')
G.add_edge('Ian','Zhi')
G.add_edge('Zhi','Ian')

# Setting node positions  
pos={'Tyler':(0,1),
     'John':(0,-1),
     'Euclid':(1,2),
     'Stefano':(1,0),
     'Bharani':(1,-2),
     'Ian':(4,0),
     'Karim':(2,-1),
     'Zhi':(5,0),
     'Mustafa':(2,1),
     'Tage':(3,0)}

## Setting parameter for nodes, edges and labels

# nodes
nx.draw_networkx_nodes(G,pos,node_size=400, node_color='g')

# edges
nx.draw_networkx_edges(G,pos,alpha=0.5,width=10)

# labels
nx.draw_networkx_labels(G,pos,font_size=15,font_family='Comic Sans MS')

# Final Plot
plt.axis('off')
plt.show() 
```
